# Lasso REgression

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge, Lasso
from sklearn.linear_model import RidgeCV, LassoCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score
```

#### Veri Seti

```python
#ridge regresyonu için kullanılacak olan veri seti beyzbol oyuncularının içerdiği beyzbol veri seti
df = pd.read_csv("beyzbol_veri.csv")

df = df.dropna()   #Şuanda veri önişleme konu ile alakalı olmadığından işleme alınmamış eksik değerler uçurulmuştur.

dms = pd.get_dummies(df[["League", "Division", "NewLeague"]])  #One-hat dönüşümü degisken_standart_donusumu.jpynb

y = df["Salary"]  #oyuncuların değerleri ile tahmin yapılacak. Bağımlı değişken

X_ = df.drop(["Salary","League", "Division", "NewLeague"], axis = 1).astype("float64") 
# df içinden league division ve newleague değişkenleri uçurarak one-hat dönüşümü yapılan bu değişkenler kullanılacak.

X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis = 1) #bağımsız değişkenler.

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.25, random_state = 42)
```

## Model kurma

```python
lasso_model = Lasso().fit(X_train, y_train)  #lamda suanda alınmamıştır.
y_test_pred = lasso_model.predict(X_test)
RMSE=np.sqrt(mean_squared_error(y_test, y_test_pred))
```

## Model Tuning  

```python
lamdalar = 10*np.linspace(10,-2,100)*0.5
```

```python
lasso_model_CV = LassoCV( cv= 10 , max_iter = 10).fit(X_train, y_train)
```

```python
lasso_model_CV.alpha_  #en uygun alpha değerini kendi belirledi.
```

```python
lass_model_CV = LassoCV(alphas = lamdalar, cv= 10, max_iter = 1000).fit(X_train, y_train)
```

```python
lass_model_CV.alpha_  # en uygun lamdayı lamda listesi arasından seçildi.
```

```python
#Son olarak model kurma uygun alpha değerini aldıktan sonra model kurma
```

```python
lass_tuned = Lasso(alpha = lass_model_CV.alpha_).fit(X_train, y_train)
```

```python
y_pred = lass_tuned.predict(X_test)
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
RMSE
```

```python
# Bağımsız değişkenlerce bağımlı değişkendeki değişikliğin açıklanma yüzdesi olan r2_score'u
```

```python
r2_score(y_test, y_pred)  #Açıklanabilirlik.
```
