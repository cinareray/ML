# **DEĞİŞKEN STANDARDİZYONU ( VERİ STANDARDİZYONU)**

* DEĞİŞKEN DÖNÜŞTÜRME: VERİ SETİNDEKİ BİLGİ BOZAR BAZENDE TEMSİL ŞEKİLİNİ DEĞİTİRİR(0-1 GİBİ)
* DEĞİŞKEN STANDARTLAŞTIRMAK : VERİ SETİNDE BULUNAN BİLGİYİ ÖZÜTÜNÜ(VARYASINI ORTALAMASINI) BOZMADAN BİLGİLERİ BELİRLİ BİR STANDARDA GETİRİR.

```python
import numpy as np
import pandas as pd
V1 = np.array([10 , 55, 154, 18, 84, 6])
V2 = np.array([643 , 531, 51, 13, 14, 12])
V3 = np.array([7 , 735, 364, 68, 54, 41])
df = pd.DataFrame({"V1": V1,
                  "V2": V2,
                  "V3":V3})
df = df.astype("float64")  #verileri floata çevirme
df
```

## STANDARDİZASYON

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)  # sklearn kutuphanesinden preprocessing fonksiyonu ile veri bilgileri standardizosyona yapılabilir
```

## Normalizasyon

0-1 arasında bilgileri standartlandırır.

```python
preprocessing.normalize(df)  #0-1 arasında standartlandırıldı.
```

## MİN-MAX standartlandırma

Min ve Max olarak girilen değerler arasında standartlandırma yapılır.

```python
scale = preprocessing.MinMaxScaler(feature_range = (1, 10))  #1-10 değer arasında standarlandırılır.
#Not : yukaradaki scale örneklendirme yapılmıştır. Aslında dönüştürme nesnesi oluşturuldu.
```

```python
scale.fit_transform(df)
```

# **DEĞİŞKEN DÖNÜŞTÜRME**

* **Değişken standardizasyonu bir değişken dönüştürmedir ancak Birbiri arasındaki fark bilinmelidir(yukarıda belirtildi)**
* **Değişken dönüştürme kategoriza edilmiş değişkenlerin numeric ifadelere dönüştürülmesi olarak düşünülebilir**

```python
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

## 0-1 numerik dönüştürülmesi

* **tips data setinin "sex" değişkeninin 2 adet kategorisi vardır. Erkek ve kadın olarak. 0 ve 1 olarak değişken dönüştürülebilir.**
* **Dikkat edilmesi gereken dönüştürüldükten sonra hangi kategoriye 1 hangisine 0 atandığı iyi anlaşılmalıdır.**

```python
from sklearn.preprocessing import LabelEncoder

lbe = LabelEncoder()  # labelencoder fonksiyonu bir dönüştürme nesnesi oalrak ele alındı. DÖNÜŞTÜRÜCÜ daha sonra kullanılacak.
```

```python
lbe.fit_transform(df["sex"])
```

```python
df["new_sex"] = lbe.fit_transform(df["sex"])  # dönüştürme dataframe "new_sex" olarak eklendi.
```

```python
df[["sex","new_sex"]].head() #görüldüğü üzere Female = 0, Male = 1 olarak dönüştürümüştür.
```

## 1 VE DİĞERLERİ 0 = SEÇİLEN DEĞİŞKENİN BİR KATEGORİSİ 1 DİĞERLERİ 0 OLARAN DÖNÜŞTÜRÜLMESİ.

```python
import numpy as np
df.head()
```

```python
#aşağıdaki kod: where operatörü koşul çalıştırır. df["day"].str.contains("Sun") ise df'nin "day" değişkenin str olarak ele al
#içerisinde "Sun" var ise (ilk parametre .., 1 , 0)) 1 diğer günler için 0 değeri dönüştürülecektir.  
```

```python
df["new_day"] = np.where( df["day"].str.contains("Sun"), 1, 0)
```

```python
df[["day","new_day"]]
```

## One-Hat Dönüşümü ve dummy

* **Çok kategorili bir değişkeni one-hat değişkeni ile dönüştürme yapılırsa, kategöriler dataframe'e değişken olarak girecektir.**
* **Dummy Değişken Tuzağı : 2 kategorili bir değişken One-hat ile dönüştürülür ise iki ayrı değişken daha eklenecektir ve bu iki değişken birbiri ile tamamen bağlı olacaktır yani ikiside aynı bilgiyi verecektir. Bundan kaçınmanın yolu kategori sayısından 1 eksik şekilde dönüştürülmesi gerekir.**

```python
import pandas as pd 
import numpy as np
import seaborn as sns 
df = sns.load_dataset("tips")
df.head()
```

### **One-Hat dönüşümü "sex" değişkenine uygulanırsa** 

```python
pd.get_dummies(df, columns = ["sex"], prefix = ["sex"]).head()
# get_dummies pandas ile uyumlu. df dataframına "sex" değişkenine uygula prefix ise olucak değişkenlerin hangi 
#değişkenden meydanaya gelidiğini ifade etmekte
```

* **_Görüldüğü üzere sex_Male ve sex_Female değişkenleri oluştu ve bu değişkenler aynı bilgiyi aktarmaktadır._**

### One-Hat dönüşümü Çok kategorili olan "day" değişkenine uygulanırsa.

```python
pd.get_dummies(df, columns = ["day"], prefix = ["day"])
```

* **_Göürldüğü üzere "day" değişkeni dataframeden silinip kategörileri değişken olarak eklenmekte. Thusday olan kategöri 1 diğerleri 0'dır.**
